```python
import pandas as pd # подключим библиотеки
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

2 загрузите в переменную df набор данных, содержащийся в файле «BankChurnersRAW.csv», используйте требуемую кодировку подсчитайте число уникальных значений по признаку «Card_Category»

```python
data = pd.read_csv('BankChurnersRAW.csv', encoding = 'cp1251', sep = ';')
data
```

```python
data['Card_Category'].value_counts()
```

```python
data['Churn'] = data['Churn'].astype('int64')
```

3 отсортируйте df по возрастанию переменной, соответствующей размеру лимита по кредиту; какой номер клиента, получившего самый крупный лимит?

```python
data = data.sort_values(by='CreditLimit', ascending=False)
data
```

```python
data.iloc[0,0]
```

4 рассчитайте среднее значение кредитного лимита по всем клиентам

```python
data['CreditLimit'].mean()
```

5 рассчитайте среднее значение кредитного лимита по тем клиентам, которые покинули банк (Churn = 1)

```python
data[data['Churn'] == 1]['CreditLimit'].mean()
```

6 определить максимальный "коэффициент использования кредита" для не-лояльных клиентов, у которых более 2 «dependents»

```python
data[(data['Churn'] == 1) & (data['Dependents'] > 2)]['AvgUtilizationRatio'].max()
```

7 Провести группировку по типу кредитной карты и рассчитать для каждой группы, максимальное, минимальное, среднее и медианное значение кредитного лимита

```python
data.groupby(['Card_Category'])['CreditLimit'].agg([np.mean, np.std, np.min, np.max])
```

8 построить сводную таблицу (pivot table) для кредитного лимита в зависимости от типа кредитной карты

```python
data.pivot_table(values = ['CreditLimit'], index = ['Card_Category'], aggfunc=('min','max','mean','sum'))
```

9 построить crosstab для оттока (churn) и числа иждивенцев (dependents), сделать выводы

```python
pd.crosstab(data['Churn'], data['Dependents'])
```

```python
pd.concat([pd.crosstab(data['Churn'], data['Dependents'],
           normalize=True),
           pd.crosstab(data['Churn'], data['Dependents'])], axis=1)
```

постройте горизонтальный график bar среднего коэффициента использования кредита (AvgUtilizationRatio) в зависимости от вида карты клиента (Blue, Silver и т.д.). Задайте необходимые подписи и заголовок рисунка

```python
sns.barplot(data=data, x='AvgUtilizationRatio', y='Card_Category')

# Установка заголовка и меток
plt.title('Коэффициента использования кредита в зависимости от типа карты')
plt.xlabel('Коэффициент использования кредита')
plt.ylabel('Тип карты')

# Показываем график
plt.show()
```
