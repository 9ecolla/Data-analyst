```python
import pandas as pd
data = pd.read_csv('var3.csv', encoding = 'cp1251', index_col=0)
data.head()
```

Балансировка выборки с помощью SMOTE

```python
#!pip install imblearn
```

```python
# если библиотека не установлена, раскомментируйте  и выполните

import imblearn
```

```python
X = data.loc[:, data.columns != 'churn']
y = data.loc[:, data.columns == 'churn']
```

```python
y
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split # подключаем функцию для разделения данных

os = SMOTE(random_state=0)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train
```

```python
columns = X_train.columns

os_data_X, os_data_y=os.fit_resample(X_train, y_train)
```

```python
os_data_X.info()
```

```python
y
```

```python
os_data_X = pd.DataFrame(data=os_data_X, columns=columns)
os_data_y= pd.DataFrame(data=os_data_y, columns=['churn'])
# Мы можем проверить числа наших данных
print("Объем восстановленной (oversampled) выборки: ",len(os_data_X))
print("Число оставшихся клиентов ",len(os_data_y[os_data_y['churn']==0]))
print("Число ушедших клиентов ",len(os_data_y[os_data_y['churn']==1]))
print("Доля оставшихся клиентов ",len(os_data_y[os_data_y['churn']==0])/len(os_data_X))
print("Доля ушедших клиентов ",len(os_data_y[os_data_y['churn']==1])/len(os_data_X))
```

```python
os_data_X.to_csv('os_data_X.csv', index=False)
os_data_y.to_csv('os_data_y.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
```

```python
from sklearn.model_selection import StratifiedKFold

tree = DecisionTreeClassifier(max_depth=10, random_state=3)

tree.fit(os_data_X, os_data_y.values.ravel())
```

```python
from sklearn.metrics import accuracy_score

tree_pred = tree.predict(X_test)
accuracy_score(y_test, tree_pred)
```
